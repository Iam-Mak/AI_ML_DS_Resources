# NumPy

### Creation of array

```python
my_list = [1,2,3,4,5,6]
print(my_list)
```

To cretae numpy array, we first need to import numpy package

```python
import numpy as np
```

```python
array = np.array(my_list, dtype = int)
print(array)
```

```python
print(type(array))
print(len(array))
print(array.ndim)
print(array.shape)
```

```python
array2 = array.reshape(3,2)
print(array2)
array2.shape
```

```python
array3 = array2.reshape(3,-1)
print(array3)

print(array3.ndim)
```

```python
# initializing numpy arrays from nested python lists

my_list2 = [1,2,3,4,5]
my_list3 = [2,3,4,5,6]
my_list4 = [9,7,6,8,9]

mul_arr = np.array((my_list2,my_list3,my_list4))
print(mul_arr)
print(mul_arr.shape)
```

```python
mul_arr.reshape(1,15)
```

### NumPy Arthemetics operations

```python
x = np.array([[1,2],[2,30]], dtype = np.float64)
y = np.array([[5,6],[7,8]], dtype = np.float64)
print(x)
print(y)
```

```python
print(x + y)
print(np.add(x,y))
```

```python
print(x - y)
print(np.subtract(x,y))
```

```python
print(x * y)
print(np.multiply(x,y))
```

```python
print(x.dot(y))
print(np.dot(x,y))
```

```python
print(x/y)
print(np.divide(x,y))
```
