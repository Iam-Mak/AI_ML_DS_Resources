## Measures of Central Tendency
- Measures of central tendency yield information about “particular places or locations in a group of numbers.”
- A single number to describe the characteristics of a set of data.

### Summary statistics

- **Central tendency or measures of location** - Arithmetic mean, Weighted mean, Median, Percentile.
- **Dispersion**  Skewness, Kurtosis, Range, Interquartile range,Variance, Standard score, Coefficient of variation.

### Measures of Shape

1. Skewness
- Absence of symmetry
- Extreme values in one side of a distribution
2. Kurtosis
Peakedness of a distribution
- Leptokurtic: high and thin
- Mesokurtic: normal shape
- Platykurtic: flat and spread out
3. Box and Whisker Plots
- Graphic display of a distribution
- Reveals skewness

### Finding minimum and maximum

```python
data = [1,3,4,463,2,3,6,8,9,4,254,6,72]
min(data),max(data)
```

```python
def min_and_max(data):
    min_val = min(data)
    max_val = max(data)
    return (min_val, max_val)
```

```python
min_and_max(data)
```

```python
def rangef(data):
    min_val = min(data)
    max_val = max(data)
    return (max_val - min_val)

rangef(data)
```
